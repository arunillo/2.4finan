```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, f1_score
from imblearn.over_sampling import SMOTE
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, callbacks

# Загружаем данные
df = pd.read_csv('Bank Customer Churn Prediction.csv')
print("Размер данных:", df.shape)
df.head()

# Проверка пропусков
print("\nПропущенные значения:\n", df.isnull().sum())

# Удаляем customer_id (неинформативный признак)
df.drop('customer_id', axis=1, inplace=True)

# Кодируем категориальные признаки country и gender
le_country = LabelEncoder()
le_gender = LabelEncoder()
df['country'] = le_country.fit_transform(df['country'])
df['gender'] = le_gender.fit_transform(df['gender'])

# Выделяем признаки и целевую переменную
X = df.drop('churn', axis=1)
y = df['churn']

# Удаление выбросов (IQR) – только для числовых столбцов
numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
Q1 = X[numeric_cols].quantile(0.25)
Q3 = X[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Отмечаем строки без выбросов
mask = ~((X[numeric_cols] < lower_bound) | (X[numeric_cols] > upper_bound)).any(axis=1)
X_clean = X[mask]
y_clean = y[mask]
print(f"Удалено выбросов: {len(X) - len(X_clean)}")

# Разделение на обучающий и тестовый наборы (75/25)
X_train, X_test, y_train, y_test = train_test_split(
    X_clean, y_clean, test_size=0.25, random_state=2026, stratify=y_clean
)

# Нормализация числовых признаков
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Преобразуем обратно в DataFrame для удобства (можно и без)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)

# Применяем SMOTE для балансировки классов в обучающей выборке
smote = SMOTE(random_state=2026)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)
print("Распределение классов после SMOTE:\n", pd.Series(y_train_res).value_counts())

# Построение и обучение моделей
# 1. Дерево решений (Decision Tree) с подбором гиперпараметров

# Параметры для GridSearchCV
param_grid_dt = {
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'class_weight': ['balanced', None]
}

dt = DecisionTreeClassifier(random_state=2026)
grid_dt = GridSearchCV(dt, param_grid_dt, cv=5, scoring='f1', n_jobs=-1)
grid_dt.fit(X_train_res, y_train_res)

best_dt = grid_dt.best_estimator_
print(f"Лучшие параметры Decision Tree: {grid_dt.best_params_}")
print(f"Лучшее F1 на кросс-валидации: {grid_dt.best_score_:.4f}")

# 2. Ансамблевый классификатор (Random Forest) с подбором гиперпараметров

param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 15, None],
    'min_samples_split': [2, 5],
    'class_weight': ['balanced', None]
}

rf = RandomForestClassifier(random_state=2026, n_jobs=-1)
grid_rf = GridSearchCV(rf, param_grid_rf, cv=5, scoring='f1', n_jobs=-1)
grid_rf.fit(X_train_res, y_train_res)

best_rf = grid_rf.best_estimator_
print(f"Лучшие параметры Random Forest: {grid_rf.best_params_}")
print(f"Лучшее F1 на кросс-валидации: {grid_rf.best_score_:.4f}")

# 3. Нейросетевой классификатор (MLP)

# Создадим модель Keras
model_nn = keras.Sequential([
    layers.Input(shape=(X_train_res.shape[1],)),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model_nn.compile(optimizer='adam',
                 loss='binary_crossentropy',
                 metrics=[keras.metrics.F1Score(threshold=0.5, name='f1')])

# Ранняя остановка
early_stop = callbacks.EarlyStopping(monitor='val_f1', mode='max', patience=10, restore_best_weights=True)

# Обучение
history = model_nn.fit(
    X_train_res, y_train_res,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

"""## Оценка моделей на тестовом наборе

Вывод матриц ошибок, отчетов о классификации и ROC-кривых для всех трёх моделей.
"""

def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)
    if hasattr(model, 'predict_proba'):
        y_proba = model.predict_proba(X_test)[:, 1]
    else:
        y_proba = model.predict(X_test)  # для нейронной сети уже вероятности

    cm = confusion_matrix(y_test, y_pred)
    cr = classification_report(y_test, y_pred, target_names=['No Churn', 'Churn'])
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)

    print(f"\n===== {model_name} =====")
    print("Confusion Matrix:\n", cm)
    print("\nClassification Report:\n", cr)
    print(f"ROC AUC: {roc_auc:.4f}")
    return fpr, tpr, roc_auc

# Для дерева решений
fpr_dt, tpr_dt, auc_dt = evaluate_model(best_dt, X_test_scaled, y_test, "Decision Tree")

# Для случайного леса
fpr_rf, tpr_rf, auc_rf = evaluate_model(best_rf, X_test_scaled, y_test, "Random Forest")

# Для нейронной сети
y_pred_nn = (model_nn.predict(X_test_scaled) > 0.5).astype(int)
y_proba_nn = model_nn.predict(X_test_scaled).flatten()
cm_nn = confusion_matrix(y_test, y_pred_nn)
cr_nn = classification_report(y_test, y_pred_nn, target_names=['No Churn', 'Churn'])
fpr_nn, tpr_nn, _ = roc_curve(y_test, y_proba_nn)
auc_nn = auc(fpr_nn, tpr_nn)

print("\n===== Neural Network =====")
print("Confusion Matrix:\n", cm_nn)
print("\nClassification Report:\n", cr_nn)
print(f"ROC AUC: {auc_nn:.4f}")

# Визуализация ROC-кривых всех моделей на одном графике

plt.figure(figsize=(8, 6))
plt.plot(fpr_dt, tpr_dt, label=f'Decision Tree (AUC = {auc_dt:.3f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc_rf:.3f})')
plt.plot(fpr_nn, tpr_nn, label=f'Neural Network (AUC = {auc_nn:.3f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривые классификаторов')
plt.legend()
plt.grid(True)
plt.show()

# Выбор лучшей модели по F1 для класса churn=1

#Вычислим F1-score для класса "1" на тестовой выборке для каждой модели и определим лучшую.

f1_dt = f1_score(y_test, best_dt.predict(X_test_scaled), pos_label=1)
f1_rf = f1_score(y_test, best_rf.predict(X_test_scaled), pos_label=1)
f1_nn = f1_score(y_test, y_pred_nn, pos_label=1)

print(f"F1 для класса churn=1:")
print(f"Decision Tree : {f1_dt:.4f}")
print(f"Random Forest : {f1_rf:.4f}")
print(f"Neural Network: {f1_nn:.4f}")

best_f1 = max(f1_dt, f1_rf, f1_nn)
best_model_name = ['Decision Tree', 'Random Forest', 'Neural Network'][np.argmax([f1_dt, f1_rf, f1_nn])]
print(f"\nЛучшая модель: {best_model_name} с F1 = {best_f1:.4f}")
print(f"Ответ: Значение показателя F1 для класса churn=1 для лучшей модели = {best_f1:.4f}")
```
